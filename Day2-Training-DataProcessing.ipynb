```python
import pandas as pd

customerdf=pd.read_csv('Customer-Churn-Records.csv')

display(customerdf)
print(customerdf.shape[0])
```

```python
customerdf["Surname"]=customerdf["Surname"].str.upper()# update all records to lower case

display(customerdf)
```

```python
customerdf.dropna()#drop all null rows across any column

customerdf.drop_duplicates()

display(customerdf)
```

```python
for i in customerdf.index:
    if customerdf.loc[i,"IsActiveMember"] < 1:
        customerdf.drop(i,inplace=True)
#display(customerdf)
```
